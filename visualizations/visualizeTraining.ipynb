```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Reward**

```python
model1Reward = pd.read_csv('trainData/model1_train_reward.csv')
model2Reward = pd.read_csv('trainData/model2_train_reward.csv')
model3Reward = pd.read_csv('trainData/model3_train_reward.csv')
```

```python
rewardData = []

for r1, r2, r3 in zip(model1Reward.iterrows(), model2Reward.iterrows(), model3Reward.iterrows()):
    rewardData.append([r1[1]['Step'], r1[1]['Value'], r2[1]['Value'], r3[1]['Value']])
rewardData = np.array(rewardData)
```

```python
sns.set_style('darkgrid')
sns.despine()

maxStep = 1000

plt.plot(rewardData[:,0][:maxStep], rewardData[:,1][:maxStep])
plt.plot(rewardData[:,0][:maxStep], rewardData[:,2][:maxStep])
plt.plot(rewardData[:,0][:maxStep], rewardData[:,3][:maxStep])
plt.legend(['=0', '=0.1', '=1.0'], frameon=False)
plt.xlabel("Timestep")
plt.ylabel("Mean Reward")
plt.title("Reward Progression over Training")
plt.show()
```
